```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("athletes.csv")
```

```python
df
```

# Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Height'].fillna(df['Height'].mean(), inplace=True)
df['Weight'].fillna(df['Weight'].mean(), inplace=True)
```

```python
df['Medal'].fillna('Unknown', inplace=True)
```

```python
df.head(10)
```

# EDA

```python
# Gold Medals
```

```python
df['Gold Medal'] = df['Medal'] == 'Gold'
gold_medals_count = df.groupby('Team')['Gold Medal'].sum().reset_index()
top_10_teams = gold_medals_count.sort_values(by='Gold Medal', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Gold Medal', data=top_10_teams, palette='viridis' , hue="Team")
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.xlabel('Team')
plt.ylabel('Number of Gold Medals')
plt.title('Number of Gold Medals by Team')
plt.show()
```

# Silver Medal

```python
df['Silver Medal'] = df['Medal'] == 'Silver'
silver_medals_count = df.groupby('Team')['Silver Medal'].sum().reset_index()
top_10_teams = silver_medals_count.sort_values(by='Silver Medal', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Silver Medal', data=top_10_teams, palette='viridis' , hue="Team")
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.xlabel('Team')
plt.ylabel('Number of Silver Medals')
plt.title('Number of Silver Medals by Team')
plt.show()
```

# Bronze Medal

```python
df['Bronze Medal'] = df['Medal'] == 'Bronze'
Bronze_medals_count = df.groupby('Team')['Bronze Medal'].sum().reset_index()
top_10_teams = Bronze_medals_count.sort_values(by='Bronze Medal', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Bronze Medal', data=top_10_teams, palette='viridis' , hue="Team")
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.xlabel('Team')
plt.ylabel('Number of Bronze Medals')
plt.title('Number of Bronze Medals by Team')
plt.show()
```

# overall Medals

```python
df['overall'] = df['Medal'].isin(['Bronze', 'Silver', 'Gold'])
overall = df.groupby('Team')['overall'].sum().reset_index()
top_10_teams = overall.sort_values(by='overall', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='overall', data=top_10_teams, palette='viridis' , hue="Team")
plt.xticks(rotation=90)  # Rotate x-axis labels if needed
plt.xlabel('Team')
plt.ylabel('Number of Medals')
plt.title('Number of  Medals by Team')
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame setup
# df = pd.read_csv('your_data.csv')

# Group by 'Year' and 'Country' and count the number of medals won
medal_counts_per_year = df.groupby(['Year', 'Team'])['Medal'].count().reset_index()

# Rename columns for clarity
medal_counts_per_year.columns = ['Year', 'Team', 'Number of Medals']

# Sort by year and number of medals (descending order)
medal_counts_per_year = medal_counts_per_year.sort_values(by=['Year', 'Number of Medals'], ascending=[True, False])

# Filter to show top 10 countries by number of medals each year
top_10_countries_per_year = medal_counts_per_year.groupby('Year').apply(lambda x: x.head(10)).reset_index(drop=True)

# Create the plot
plt.figure(figsize=(14, 8))
sns.barplot(x='Year', y='Number of Medals', hue='Team', data=top_10_countries_per_year, palette='tab10')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Number of Medals Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Country')

# Show the plot
plt.show()
```

```python
# Get the top 10 sports
top_10_sports = df['Sport'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 sports
top_10_sports_df = df[df['Sport'].isin(top_10_sports)]

# Count the number of medals for each sport in the top 10
top_10_sports_counts = top_10_sports_df['Sport'].value_counts().reset_index()
top_10_sports_counts.columns = ['Sport', 'Number of Medals']

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Sport', y='Number of Medals', data=top_10_sports_counts, palette='viridis' , hue="Sport")

# Rotate x-axis labels if needed
plt.xticks(rotation=45)
plt.title('Top 10 Sports by Number of Medals')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
```
